```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100 #number of nodes
M = 10 # number of communities 
G = nx.random_partition_graph([int(N/M)]*M,0.8,0.005)

edgeswap = N**2
H = G.copy()
nx.double_edge_swap(H,nswap=edgeswap, max_tries=edgeswap*10)
```

```python

position = nx.spring_layout(G)
nx.draw(G, pos = position, node_color='b', width=1.0, node_size = 10)
```

```python

position = nx.spring_layout(G)
nx.draw(H, pos = position, node_color='b', width=1.0, node_size = 10)
```

```python
def signalupdate(G, node):
    state_perturbation = 0
    signal_propagation = 0
    for neigh in G.neighbors(node):
        #state_perturbation = state_perturbation + G.nodes[neigh]['state_t-1']**2 + G.nodes[neigh]['state_t-1']*G.nodes[neigh]['state_t-2']
        state_perturbation = state_perturbation + (G.nodes[neigh]['state_t']- G.nodes[neigh]['state_t-1'])**2
        #signal_propagation = signal_propagation + G.nodes[neigh]['signal_t-1']- damper
        signal_propagation = signal_propagation + G.nodes[neigh]['signal_t-1']*(G.nodes[neigh]['signal_t-1'] - G.nodes[neigh]['signal_t-2'])- damper
    if state_perturbation > 0 or signal_propagation > 0:
        G.nodes[node]['signal_t'] = 1
        return 1
    else:
        G.nodes[node]['signal_t'] = 0
        return 0
```

```python
def stateupdate(G,node):
    
    if G.nodes[node]['signal_t'] == 0:
        return 0
    elif G.nodes[node]['signal_t'] == 1:
        for neigh in H.neighbors(node):
            if neigh not in G.neighbors(node) and G.nodes[neigh]['signal_t-1']  == 0:
                G.nodes[node]['state_t'] = 0
                return 0
            else:
                S = 0
                for neigh in G.neighbors(node):
                    for neigh_2 in H.neighbors(neigh):
                        if neigh_2 not in G.neighbors(node) and neigh_2 != node and G.nodes[neigh_2]['signal_t-1']  == 0:
                            G.nodes[node]['state_t'] = 0
                            S = 1
                            break
                    break
                if S == 1:
                    return 0
                elif S ==0:
                    G.nodes[node]['state_t'] = 1
                    for neigh in G.neighbors(node):
                        G.nodes[neigh]['state_t'] = 1
                    return 1
```

```python
damper = 0.1
nx.set_node_attributes(G, 0,'signal_t')
nx.set_node_attributes(G, 0,'signal_t-1')
nx.set_node_attributes(G, 0,'signal_t-2')
nx.set_node_attributes(G, 0,'state_t-2')
nx.set_node_attributes(G, 0,'state_t-1')
nx.set_node_attributes(G, 0,'state_t')
```

```python
# initialise
trialtime = 40
infected_nodes = np.zeros([trialtime])
signal_density = np.zeros([trialtime])

heat = 0.01 # fraction of nodes to incite 
initial_rebels = set(np.random.choice(range(N), int(N*heat), replace=False))
for node in initial_rebels:
    G.nodes[node]['signal_t'] = 1
    #G.nodes[node]['state_t'] = 1
    #G.nodes[node]['state_t-1'] = 0
signal_density[0] = heat
#infected_nodes[0] = heat
```

```python


for t in range(1,trialtime):
    for node in G.nodes():
        G.nodes[node]['signal_t-2'] = G.nodes[node]['signal_t-1']
        G.nodes[node]['signal_t-1'] = G.nodes[node]['signal_t']
    for node in G.nodes():
        signalupdate(G, node)
        signal_density[t] = signal_density[t] + G.nodes[node]['signal_t']/N
    
    G.nodes[node]['state_t-2'] = G.nodes[node]['state_t-1']
    G.nodes[node]['state_t-1'] = G.nodes[node]['state_t']

    for node in G.nodes():
        stateupdate(G,node)
        infected_nodes[t] = infected_nodes[t] + G.nodes[node]['state_t']/N
    '''
    if infected_nodes[t] > 1-1/N:
        infected_nodes[range(t+1, trialtime)] = 1
        signal_nodes[range(t+1, trialtime)] = 1
        break
    '''
```

```python
window = 10
t = range(window)

plt.plot(t, infected_nodes[t], 'b-',t, signal_density[t], 'r-')
plt.ylim(0, 1)
```
